```python
# importing required libraries
import pandas as pd
import numpy as np
```

```python
# importing data as required
crd_data=pd.read_csv('tmdb_5000_credits.csv')
mv_data=pd.read_csv('tmdb_5000_movies.csv')
```

```python
# merging both datasets on common column title of movie
movie=mv_data.merge(crd_data,on='title')
```

```python
# required columns : 
# genres
# keywords
# title
# overview
# id
# cast
# crew
```

```python
dt=movie[['genres','keywords','title','overview','id','cast','crew']]
```

```python
dt.head()
```

```python
# checking if there are any NULL column
dt.isnull().sum()
```

```python
dt.iloc[0].genres
```

```python
# In genres column we need only genres names and not other not other alpha variables 
#The literal_eval() function is particularly useful when you want to evaluate a string containing a literal expression or data structure that was entered by a user, without risking arbitrary code execution or other security vulnerabilities. It provides a safer alternative to using eval() function, which can execute arbitrary code and therefore poses a security risk.
```

```python
import ast
def fun1(tmp):
    lst=[]
    for i in ast.literal_eval(tmp):
        lst.append(i['name'])
        
    return lst
```

```python
dt['genres']=dt['genres'].apply(fun1)
```

```python
dt.head()
```

```python
dt.iloc[0].keywords
```

```python
# same thing we need to do with kewords 
dt['keywords']=dt['keywords'].apply(fun1)
```

```python
dt.head()
```

```python
# for cast column we since there could be many actors\actress in a particular movie we need only major actors in the movie which could be the first 3 members in cast
```

```python
def fun2(tmp):
    lst=[]
    num=0
    for i in ast.literal_eval(tmp):
        if num!=3:
            lst.append(i['name'])
            num+=1
        else:
            break
    return lst
```

```python
dt['cast']=dt['cast'].apply(fun2)
```

```python
dt.head()
```

```python
# for crew column our main target could be director 
dt.iloc[0].crew
```

```python
def fun3(tmp):
    lst=[]
    for i in ast.literal_eval(tmp):
        if i['job']=='Director':
            lst.append(i['name'])
            break
    return lst
```

```python
dt['crew']=dt['crew'].apply(fun3)
```

```python
dt.head()
```

```python
# finally we will consider ovewview column
dt.iloc[0].overview
```

```python
# for overview column we will be taking each word of every string as a single object for each row
dt['overview']=dt['overview'].apply(lambda x:x.split())
```

```python
dt.iloc[0].overview
```

```python
# for genres,keywords,cast,crew we have to remove the space in between 
# otherwise different objects would behave similarly
# Example : For recommending movies of Mukesh sharma our model could also suggest movies of  Mukesh Verma (say name of directors) 
```

```python
dt['keywords']=dt['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
dt['crew']=dt['crew'].apply(lambda x:[i.replace(" ","") for i in x])
dt['cast']=dt['cast'].apply(lambda x:[i.replace(" ","") for i in x])
dt['genres']=dt['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
dt.head()
```

```python
dt['tags']=dt['overview']+dt['genres']+dt['keywords']+dt['cast']+dt['crew']
```

```python
dt.head()
```

```python
mdt=dt[['id','title','tags']]
```

```python
mdt.head()
```

```python
mdt['tags']=mdt['tags'].apply(lambda x:" ".join(x))
```

```python
mdt.head()
```

```python
mdt.iloc[0].tags
```

```python
mdt['tags']=mdt['tags'].apply(lambda x:x.lower())
```

```python
mdt.head()
```

```python
mdt['tags'][0]
```

```python
mdt['tags'][1]
```

```python
# our main goal for finding the similarity between the movies is to find the similarity between these tags
# We will be extracting 10,000 most frequent words from the tags of all the movies
# for converting the tags into vectors we will be comparing how many times each words from 10000 most frequent words appear in a particular tag of movie
# we also have to check not to exceed more words in most frequent words other time complexity would be very high
```

```python
# in our most frequent words similar words may be repeating , example love,loving,loves
# stemming is applied to remove those similar words from our most frequent words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(tmp):
    lst=[]
    for i in tmp.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
mdt['tags']=mdt['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
vectors=cv.fit_transform(mdt['tags']).toarray()
```

```python
# frequent words
cv.get_feature_names()
```

```python
vectors
```

```python
# since higher dimension restricts using euclidean distance we will be using cosine distance as a measure of similarity between the tags of movies with one another
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
# here similarity[i][j] shows similarity between ith row movie and jth row movie
```

```python
# here we are getting similarity of ith movie with jth along similarity measure 
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=mdt[mdt['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movie_list:
        print(mdt.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
